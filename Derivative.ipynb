```python
def derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

def f(x):
    return x ** 2

x_value = 2.0
derivative_value = derivative(f, x_value)

print(f"The derivative of f(x) at x = {x_value} is: {derivative_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

def f(x):
    return x ** 2

def f_prime(x):
    return 2 * x

x_values = np.linspace(-5, 5, 400)
y_values = f(x_values)
y_prime_values = f_prime(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = x^2')
plt.plot(x_values, y_prime_values, label="f'(x) = 2x")
plt.scatter([2], [4], color='red', label="Derivative at x=2")
plt.axvline(x=2, color='gray', linestyle='--', label='x = 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Function and its Derivative")
plt.legend()
plt.grid()
plt.show()
```
