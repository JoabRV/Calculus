```python
import numpy as np

# Definici√≥n de dos vectores como arrays de NumPy
vector1 = np.array([2, 3])
vector2 = np.array([1, -1])

# Suma de los dos vectores
vector_sum = vector1 + vector2

print("Vector 1:", vector1)
print("Vector 2:", vector2)
print("Suma de vectores:", vector_sum)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de dos vectores como arrays de NumPy
vector1 = np.array([2, 3])
vector2 = np.array([1, -1])

# Suma de los dos vectores
vector_sum = vector1 + vector2

# Creaci√≥n de una figura para la gr√°fica
plt.figure(figsize=(6, 6))

# Graficar los vectores
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2')
plt.quiver(0, 0, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale=1, color='green', label='Suma')

# Configurar l√≠mites y etiquetas de ejes
plt.xlim(-1, 5)
plt.ylim(-2, 4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Suma de Vectores')
plt.legend()
plt.grid()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
