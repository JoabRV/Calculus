```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x):
    return x ** 2

x_values = np.linspace(0, 4, 400)
y_values = f(x_values)

integral_result, _ = quad(f, 0, 4)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = x^2')
plt.fill_between(x_values, y_values, alpha=0.3)
plt.text(2, 6, f"Integral = {integral_result:.2f}", fontsize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Function and its Integral")
plt.legend()
plt.grid()
plt.show()
```
